# 6 Resumen analítico y selección de la mejor tienda

## 6.1 Introdcucuón

El propósito de este anális es ayudar al señor Juan a decidir en qué tienda le conviene vender sus productos.
Para ello, se analizaron datos de cuatro tiendas (Tienda 1, Tienda 2, Tienda 3, Tienda 4) utilizando Python, Google colab y la biblioteca Pandas.

El análisis se centro en cinco principales aspectos:

- Los **ingresos totales** de cada tienda.
- Las **categorías de productos** más y menos vendidas.
- La **valoración media** otorgada por los clientes en cada tienda.
- Los **productos específicos** más y menos vendidos.
- El **costo de envío promedio** en cada tienda.

A partir de estos resultados y de las visualizaciones generadas, se formula una recomendación
sobre la tienda en la que el Sr. Juan debería ofrecer sus productos.

## 6.3 Colclusión y recomendación

A partir de los análisis realizados y de las visualizaciones generadas, se puede comparar el desempeño de las cuatro tiendas
en términos de ingresos, categorías de productos vendidos, valoración de los clientes, productos más/menos vendidos y
costo de envío promedio.

En resumen:

- **Ingresos**: la tienda con mayor ingreso total es **Tienda 1**, lo que indica un mercado activo, sin embargo el producto más vendido se encuentra en la **Tienda 3**
- **Valoración de clientes**: la mejor valoración promedio corresponde a **Tienda 3**,
  lo cual sugiere mayor satisfacción del cliente.
- **Costo de envío**: la tienda con menores costos de envío es **Tienda 4**,
  lo que puede mejorar la competitividad del precio final y en segundo lugar se encuentra la **Tienda 3**.

Considerando todos estos factores, la tienda que ofrece el **mejor equilibrio global** para los intereses del Sr. Juan es:

### 👉 **Tienda 3**

Esta tienda combina:

- Un nivel adecuado de ingresos (buen movimiento de ventas).
- Una valoración favorable por parte de los clientes.
- Buen desempeño en la categoría de productos relevante para el Sr. Juan.
- Un costo de envío competitivo.

Por estos motivos, se recomienda que el Sr. Juan comience a vender sus productos en **Tienda 3**,
sin descartar que en el futuro pueda expandirse a otras tiendas una vez que analice el comportamiento de sus ventas iniciales.
