# 1 Análisis de facturación



### Ingreso total por cada tienda

En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
# Calcular ingreso total de cada tienda por separado
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

ingresos = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]
}

df_ingresos = pd.DataFrame(ingresos)
df_ingresos
```

```python
# Generación de grafico ingreso total de cada tienda

import matplotlib.pyplot as plt

def plot_bar (df, x_col, y_col, title='', x_label='', y_label=''):
    plt.figure(figsize=(8, 5))
    plt.bar(df[x_col], df[y_col], color='orange')
    plt.title(title)
    plt.xlabel(x_label if x_label else x_col)
    plt.ylabel(y_label if y_label else y_col)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()

plot_bar (
    df_ingresos,
    x_col='Tienda',
    y_col='Ingreso Total',
    title='INGRESO TOTAL POR TIENDA',
    x_label='Tienda',
    y_label='Ingreso total'
)
```

```python
import matplotlib.pyplot as plt

def plot_barh(df, x_col, y_col, title='', x_label='', y_label=''):
    plt.figure(figsize=(8, 5))
    plt.barh(df[x_col], df[y_col], color='steelblue')
    plt.title(title)
    plt.xlabel(x_label if x_label else x_col)
    plt.ylabel(y_label if y_label else y_col)
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()


plot_barh(df_ingresos, 'Tienda', 'Ingreso Total',
              title='Ingreso total por tienda (azul)',
              x_label='Tienda', y_label='Ingreso total')

```

```python
def plot_pie(df, labels_col, values_col, title=''):
    plt.figure(figsize=(6, 6))
    plt.pie(
        df[values_col],
        labels=df[labels_col],
        autopct='%1.1f%%',
        startangle=90,
        colors=['lightgreen', 'mediumseagreen', 'yellowgreen', 'darkseagreen']
    )
    plt.title(title)
    plt.tight_layout()
    plt.show()

plot_pie(df_ingresos, 'Tienda', 'Ingreso Total',
               title='Distribución del ingreso por tienda')
```

# 2 Ventas por categoria

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
# Cantidad de productos vendidos por categoría en cada tienda

cat_tienda1 = tienda['Categoría del Producto'].value_counts()

cat_tienda2 = tienda2['Categoría del Producto'].value_counts()

cat_tienda3 = tienda3['Categoría del Producto'].value_counts()

cat_tienda4 = tienda4['Categoría del Producto'].value_counts()



print("Tienda 1:")
print(cat_tienda1)

print("\nTienda 2:")
print(cat_tienda2)

print("\nTienda 3:")
print(cat_tienda3)

print("\nTienda 4:")
print(cat_tienda4)
```

```python
# Obtener la categoría más vendida y su cantidad en cada tienda
top1_t1_cat = cat_tienda1.idxmax()   # nombre de la categoría
top1_t1_cnt = cat_tienda1.max()      # cantidad

top1_t2_cat = cat_tienda2.idxmax()
top1_t2_cnt = cat_tienda2.max()

top1_t3_cat = cat_tienda3.idxmax()
top1_t3_cnt = cat_tienda3.max()

top1_t4_cat = cat_tienda4.idxmax()
top1_t4_cnt = cat_tienda4.max()

# Construir la tabla resumen
tabla_top = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Categoría más vendida': [top1_t1_cat, top1_t2_cat, top1_t3_cat, top1_t4_cat],
    'Cantidad vendida': [top1_t1_cnt, top1_t2_cnt, top1_t3_cnt, top1_t4_cnt]
})

tabla_top
```

```python
import matplotlib.pyplot as plt

# Gráfico 1 – Barras: categoría más vendida por tienda
plt.figure(figsize=(8, 5))
plt.bar(tabla_top['Tienda'], tabla_top['Cantidad vendida'], color='orange')
plt.title('Categoría más vendida por tienda (Muebles)')
plt.xlabel('Tienda')
plt.ylabel('Cantidad vendida')
plt.tight_layout()
plt.show()

# Gráfico 2 – Pastel: distribución de categorías en la Tienda 1
plt.figure(figsize=(7, 7))
plt.pie(
    cat_tienda1.values,
    labels=cat_tienda1.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['gold', 'lightskyblue', 'lightcoral', 'palegreen', 'plum', 'khaki']
)
plt.title('Distribución de categorías – Tienda 1')
plt.tight_layout()
plt.show()


# ==========================
# Gráfico 3 – Barras agrupadas: categorías por tienda
# ==========================
# Unir los conteos de cada tienda en un solo DataFrame
df_categorias = pd.DataFrame({
    'Tienda 1': cat_tienda1,
    'Tienda 2': cat_tienda2,
    'Tienda 3': cat_tienda3,
    'Tienda 4': cat_tienda4
}).fillna(0)

plt.figure(figsize=(12, 6))
df_categorias.plot(
    kind='bar',
    figsize=(12, 6),
    color=['steelblue', 'salmon', 'seagreen', 'mediumorchid']
)

plt.title('Cantidad de productos vendidos por categoría y por tienda')
plt.xlabel('Categoría del producto')
plt.ylabel('Cantidad vendida')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Tienda')
plt.tight_layout()
plt.show()
```

# 3 Calificación promedio de la tienda


En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

```python
valoracion_tienda1 = tienda['Calificación'].mean()
valoracion_tienda2 = tienda2['Calificación'].mean()
valoracion_tienda3 = tienda3['Calificación'].mean()
valoracion_tienda4 = tienda4['Calificación'].mean()

tabla_valoracion = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Valoración media': [valoracion_tienda1, valoracion_tienda2, valoracion_tienda3, valoracion_tienda4]})
tabla_valoracion
```

```python
import matplotlib.pyplot as plt

# ==========================
# Gráfico 1 – Barras verticales (promedio por tienda)
# ==========================
plt.figure(figsize=(8, 5))
plt.bar(tabla_valoracion['Tienda'], tabla_valoracion['Valoración media'], color='teal')
plt.title('Valoración media por tienda')
plt.xlabel('Tienda')
plt.ylabel('Valoración media')
plt.ylim(0, 5)  # asumiendo escala 1–5
plt.tight_layout()
plt.show()


# ==========================
# Gráfico 2 – Barras horizontales
# ==========================
plt.figure(figsize=(8, 5))
plt.barh(tabla_valoracion['Tienda'], tabla_valoracion['Valoración media'], color='mediumorchid')
plt.title('Valoración media por tienda (barras horizontales)')
plt.xlabel('Valoración media')
plt.ylabel('Tienda')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()


# ==========================
# Gráfico 3 – Línea con marcadores
# ==========================
plt.figure(figsize=(8, 5))
plt.plot(
    tabla_valoracion['Tienda'],
    tabla_valoracion['Valoración media'],
    marker='o',
    linestyle='-',
    color='darkorange'
)
plt.title('Comparación de valoración media entre tiendas')
plt.xlabel('Tienda')
plt.ylabel('Valoración media')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# 4 Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
pro_menos_ven_t1 = tienda['Producto'].value_counts().idxmin()
pro_menos_ven_t2 = tienda2['Producto'].value_counts().idxmin()
pro_menos_ven_t3 = tienda3['Producto'].value_counts().idxmin()
pro_menos_ven_t4 = tienda4['Producto'].value_counts().idxmin()

pro_mas_ven_t1 = tienda['Producto'].value_counts().idxmax()
pro_mas_ven_t2 = tienda2['Producto'].value_counts().idxmax()
pro_mas_ven_t3 = tienda3['Producto'].value_counts().idxmax()
pro_mas_ven_t4 = tienda4['Producto'].value_counts().idxmax()


tabla_valoracion2 = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Producto más vendido': [pro_mas_ven_t1, pro_mas_ven_t2, pro_mas_ven_t3, pro_mas_ven_t4]
    , 'Producto menos vendido': [pro_menos_ven_t1, pro_menos_ven_t2, pro_menos_ven_t3, pro_menos_ven_t4]})
tabla_valoracion2

```

```python
# Creación de funciones
df_mas = pd.DataFrame({
    'Tienda': ['Tienda 1 \n (Microondas)', 'Tienda 2 \n(Iniciación en programación)', 'Tienda 3 \n(Kit de bancas)', 'Tienda 4\n(Cama box)'],
    'Producto': [pro_mas_ven_t1, pro_mas_ven_t2, pro_mas_ven_t3, pro_mas_ven_t4],
    'Cantidad': [cat_tienda1.max(), cat_tienda2.max(), cat_tienda3.max(), cat_tienda4.max()]
 })

df_menos = pd.DataFrame({
    'Tienda': ['Tienda 1  \n(Auriculares con microfono)', 'Tienda 2  \n(Juego de mesa)', 'Tienda 3  \n(Bloques de construción)', 'Tienda 4  \n(Gutarra eléctrica)'],
    'Producto': [pro_menos_ven_t1, pro_menos_ven_t2, pro_menos_ven_t3, pro_menos_ven_t4],
    'Cantidad': [cat_tienda1.min(), cat_tienda2.min(), cat_tienda3.min(), cat_tienda4.min()]
})

# Gráfico 1 - Producto más vendido
plt.figure(figsize=(9, 5))
plt.bar(df_mas['Tienda'], df_mas['Cantidad'], color='cornflowerblue')
plt.title('Producto más vendido por tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad vendida')
plt.tight_layout()
plt.show()

# Gráfico 2 – Barras horizontales: productos menos vendidos por tienda
plt.figure(figsize=(9, 5))
plt.barh(df_menos['Tienda'], df_menos['Cantidad'], color='lightcoral')
plt.title('Producto menos vendido por tienda')
plt.xlabel('Cantidad vendida')
plt.ylabel('Tienda')
plt.tight_layout()
plt.show()

# Gráfico 3 – Barras agrupadas: más vs menos vendido por tienda
import numpy as np

tiendas_labels = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
x = np.arange(len(tiendas_labels))  # posiciones en el eje x
width = 0.35  # ancho de cada barra

plt.figure(figsize=(12, 6))
bars_mas = plt.bar(x - width/2, df_mas['Cantidad'], width, label='Más vendido', color='seagreen')
bars_menos = plt.bar(x + width/2, df_menos['Cantidad'], width, label='Menos vendido', color='orange')

plt.xticks(x, tiendas_labels)
plt.title('Comparación: producto más vs menos vendido por tienda')
plt.xlabel('Tienda')
plt.ylabel('Cantidad vendida')
plt.legend()

# Añadir nombre del producto más vendido sobre su barra
for i, bar in enumerate(bars_mas):
    height = bar.get_height()
    producto = df_mas.loc[i, 'Producto']
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        producto,
        ha='center',
        va='bottom',
        rotation=45,
        fontsize=8
    )

# Añadir nombre del producto menos vendido sobre su barra
for i, bar in enumerate(bars_menos):
    height = bar.get_height()
    producto = df_menos.loc[i, 'Producto']
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        height + 0.5,
        producto,
        ha='center',
        va='bottom',
        rotation=45,
        fontsize=8
    )

plt.tight_layout()
plt.show()
```

# 5 Envío promedio por tienda

En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.  Sabiendo que el cliente paga el envío.

```python
costo_envio_t1 = tienda['Costo de envío'].mean()
costo_envio_t2 = tienda2['Costo de envío'].mean()
costo_envio_t3 = tienda3['Costo de envío'].mean()
costo_envio_t4 = tienda4['Costo de envío'].mean()

tabla_valoracion3 = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo de envio promedio': [costo_envio_t1, costo_envio_t2, costo_envio_t3, costo_envio_t4]})
tabla_valoracion3
```

```python
import matplotlib.pyplot as plt

# Gráfico 1 – Barras verticales
plt.figure(figsize=(8, 5))
plt.bar(tabla_valoracion3['Tienda'],
        tabla_valoracion3['Costo de envio promedio'],
        color='skyblue')
plt.title('Costo de envío promedio por tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de envío promedio')
plt.tight_layout()
plt.show()

# Gráfico 2 – Barras horizontales
plt.figure(figsize=(8, 5))
plt.barh(tabla_valoracion3['Tienda'],
         tabla_valoracion3['Costo de envio promedio'],
         color='tomato')
plt.title('Costo de envío promedio por tienda')
plt.xlabel('Costo de envío promedio')
plt.ylabel('Tienda')
plt.tight_layout()
plt.show()

# Gráfico 3 – Gráfico de pastel
plt.figure(figsize=(6, 6))
plt.pie(
    tabla_valoracion3['Costo de envio promedio'],
    labels=tabla_valoracion3['Tienda'],
    autopct='%1.1f%%',
    startangle=90,
    colors=['gold', 'lightgreen', 'lightskyblue', 'plum']
)
plt.title('Proporción del costo de envío promedio por tienda')
plt.tight_layout()
plt.show()
```
